```python
#GET list_of_autobahn
endpoint = "https://verkehr.autobahn.de/o/autobahn/"
response = requests.get(endpoint).json()
print(response)

#write response in file
with open("list_of_autobahn.json","w") as write_file:
    json.dump(response, write_file)
```
