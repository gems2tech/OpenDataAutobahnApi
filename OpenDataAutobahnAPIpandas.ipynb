```python
#Get road works on autbahn A5
with open("filesJson/road_works_autobahn_A5.json","r") as read_file:
    data = json.load(read_file)
print(json.dumps(data, indent=3))
```

```python
#pick out info needed
pickedDataDict = {}
pickedDataDict["title"] = data["title"]
pickedDataDict["subtitle"] = data["subtitle"]
pickedDataDict["identifier"] = data["identifier"]
pickedDataDict["coordinate-lat"] = data["coordinate"]["lat"]
pickedDataDict["coordinate-long"] = data["coordinate"]["long"]
pickedDataDict["description-Beginn"] = data["description"][0]
pickedDataDict["description-Ende"] = data["description"][1]
pickedDataDict["description-Location"] = data["description"][3]
pickedDataDict["description-Text"] = data["description"][5]
```

```python
pd. set_option('display.max_colwidth', 199)
pd.DataFrame.from_dict(pickedDataDict, orient='index').transpose()
```

```python
pd. set_option('display.max_colwidth', 199)
pd.DataFrame.from_dict(pickedDataDict, orient='index').transpose()
```
